En caso de que la matriz Jacobiana no sea constante como en el notebook anterior, basta con identificar a `jacob` con una funci贸n y evaluarla en el punto correspondiente. De esta manera en la l铆nea `8` de la la siguiente funci贸n, `Euler_mejorado_J`, se observa la evaluaci贸n de la matriz Jacobiana en cada paso de integraci贸n.

```python
def Euler_mejorado_J(edo,p_ini,t,jacob):
    # Este solo tiene J y g sin dependencia temporal.
    sol = zeros((len(t),len(p_ini)))
    sol[0,:] = p_ini
    delta = t[1]-t[0]
    for i in range(0,len(t)-1):
        sol[i+1,:] = sol[i,:] + delta*edo(sol[i,:],t[i])
        sol[i+1,:] += 0.5*(delta**2)*dot(jacob(sol[i,:],t[i]),edo(sol[i,:],t[i])) # ojo sistema autonomo
    return sol
```

### Modelo log铆stico

De esta manera, tambi茅n tenemos que programar una funci贸n que nos proporcione la matriz Jacobiana del sistema de EDO's. Veamos el ejemplo en una dimensi贸n con el modelo log铆stico:

$$
\frac{dP}{dt} = aP(1-P)
$$

donde $P=P(t)$ es la poblaci贸n, y $a$ es la tasa de crecimiento o decrecimiento.

> Sin p茅rdida de generalidad, hemos elegido una poblaci贸n cr铆tica unitaria.

En este caso, el equivalente a la matriz Jacobiana es un escalar dado por la siguiente expresi贸n:

$$
J = a(1-2P)
$$

La EDO anterior y su equivalente Jacobiana quedan representadas por las siguientes funciones:

```python
def modelo_logistico(P,t,a=0.1):
    return a*P *(1 -P)

def modelo_logistico_J(P,t,a=0.1):
    return a *(1 - 2*P)
```

La soluci贸n num茅rica para diferentes condiciones iniciales se ve de la siguiente manera.

```python
title("Modelo log铆stico para varias condiciones iniciales"),xlabel("t"),ylabel("P(t)")
tiempo = arange(0,50,0.01)
pob_ini_s = arange(0.1,1.5,0.2)
for pob_ini in pob_ini_s:
    solucion = Euler_mejorado_J(modelo_logistico,[pob_ini],tiempo,modelo_logistico_J)
    plot(tiempo,solucion,label="$P _0$ = %2.2f " %pob_ini )
legend(),grid()
show()
```

Como se puede apreciar, todas las poblaciones iniciales convergen a la poblaci贸n cr铆tica, que en este caso es 1.

### Modelo P茅ndulo

Un modelo que utiliza la matriz Jacobiana dependiente de la posici贸n es el p茅ndulo, y m谩s a煤n cuando se tiene amortiguamiento y/o forzamiento. La ecuaci贸n del p茅ndulo es la siguiente:

$$
\frac{d^2\theta}{dt^2} = -\omega_0 ^2 \sin \theta
$$

donde $\omega _0 ^2 =\sqrt{\frac{g}{\ell}}$ es la llamada frecuencia fundamental del p茅ndulo para 谩ngulos peque帽os. La ecuaci贸n anterior se puede escribir como un sistema de ecuaciones haciendo el cambio de variable $\frac{d\theta}{dt} = v_{\theta}$ y se obtiene el siguiente sistema de ecuaciones de primer orden:

$$
\frac{d\theta}{dt} = v_{\theta} \\
\frac{d v_{\theta}}{dt} = -\omega_0 ^2 \sin \theta
$$

La matriz Jacobiana de este sistema de ecuaciones es:
$$
\mathbb{J}=\left( \begin{array}{c c}
0 & 1 \\
-\omega_0 ^2 \cos \theta & 0
\end{array}\right)
$$
En este caso el sistema es aut贸nomo, es decir, no depende expl铆citamente de $t$ y ello hace que la matriz sea cuadrada. Las anteriores ecuaciones quedan representadas por las siguientes funciones:

```python
def pendulo(t_vt,t,omega_0 = 1):
    dth = t_vt[1]
    dv_th = -omega_0*sin(t_vt[0])
    return array([dth,dv_th])

def pendulo_J(t_vt,t,omega_0 = 1):
    J = zeros((2,2))
    J[0,1] = 1.0
    J[1,0] = -omega_0*cos(t_vt[0])
    return J
```

A continuaci贸n se muestra el Espacio Fase, $(\theta, v_{\theta})$, del sistema p茅ndulo para varias condiciones iniciales del tipo $(\theta _0,0)$.

```python
title("Espacio Fase"),xlabel(r"$\theta(t)$"),ylabel(r"$v_{\theta}(t)$")
tiempo = arange(0,5*pi,0.05)
th_ini_s = arange(0.1,pi-0.1,0.25)
for th_ini in th_ini_s:
    solucion = Euler_mejorado_J(pendulo,[th_ini,0],tiempo,pendulo_J)
    plot(solucion[:,0],solucion[:,1])
grid()
show()
```

Un modelo m谩s realista al anterior es introducir amortiguamiento y forzamiento. Para ello, se propone que la fricci贸n sea del tipo $ F_{fric} \propto \frac{d \theta}{dt} $ y el forzamiento $F_{forzamiento} \propto \cos(\omega_1 t)$. De esta manera la ecuaci贸n del p茅ndulo se escribe como:

$$
\frac{d ^2 \theta}{dt^2} = -\omega_0 ^2 \sin \theta -\gamma \frac{d \theta}{dt} + A \cos (\omega_1 t)
$$

donde $\gamma$ es el coeficiente de amortiguamiento que absorbe las unidades y el coeficiente de fricci贸n, $A$ es la amplitud efectiva del forzamiento y tambi茅n se ajusta a las unidades adecuadas, y $\omega _1$ es la frecuencia del forzamiento. Haciendo el mismo cambio de variable para convertir la ecuaci贸n de segundo orden a un sistema de ecuaciones de primer orden se obtiene:

$$
\frac{d\theta}{dt} = v_{\theta} \\
\frac{d v_{\theta}}{dt} = -\omega_0 ^2 \sin \theta -\gamma v_{\theta} + A \cos (\omega_1 t)
$$

La matriz Jacobiana extendida del anterior sistema de ecuaciones es:
$$
\mathbb{J}^*=\left( \begin{array}{c c c}
0 & 1 & 0\\
-\omega_0 ^2 \cos \theta & -\gamma & -A\omega_1 \sin  (\omega_1 t)
\end{array}\right)
$$

Como se puede observar en este caso se tiene una matriz no cuadrada y el vector que representa las ecuaciones diferenciales debe completarse con un n煤mero 1 en una entrada extra al final del arreglo, $\vec g ^* = (\vec g, 1)$. Las siguientes funciones representan las ecuaciones anteriores:

```python
def pendulo_ff(t_vt,t,omega_0=1.0,gamma=0.1,A_w=[0.1,0.1]):
    # t_vt = [ theta , v_theta ]
    dth = t_vt[1]
    dv_th = -omega_0*sin(t_vt[0]) - gamma*t_vt[1] + A_w[0]*cos(A_w[1]*t)
    return array([dth,dv_th,1.0])

def pendulo_ff_J(t_vt,t,omega_0=1.0,gamma=0.1,A_w=[0.1,0.1]):
    J = zeros((2,3))
    J[0,1] = 1.0
    J[1,0] = -omega_0*cos(t_vt[0])
    J[1,1] = -gamma
    J[1,2] = -A_w[0]*A_w[1]*sin(A_w[1]*t)
    return J
```

En el caso del algoritmo que utilizaremos para hacer la integraci贸n, es igual que el anterior, pero hay que tomar en cuenta que cabe la posibilidad de que el sistema no sea aut贸nomo. En las l铆neas `7`a `11`se toma la decisi贸n sobre cual de los dos algoritmos tomar en caso de que el sistema no sea aut贸nomo. En cualquiera de los dos casos, la soluci贸n que se obtiene s贸lo debe contener el n煤mero de entradas igual al n煤mero de ecuaciones diferenciales y por ello en el caso de sistemas no aut贸nomo, la suma del punto anterior termina en el pen煤ltimo punto que en python se representa con `-1`, para que la suma de los vectores tengan la misma longitud. 

```python
def Euler_mejorado_J_t(edo,p_ini,t,jacob):
    # J y g con dependencia temporal.
    sol = zeros((len(t),len(p_ini)))
    sol[0,:] = p_ini
    delta = t[1]-t[0]
    for i in range(0,len(t)-1):
        eval_edo = edo(sol[i,:],t[i])
        if len(eval_edo) == len(p_ini):
            sol[i+1,:] = sol[i,:] + delta*eval_edo
        else:
            sol[i+1,:] = sol[i,:] + delta*eval_edo[0:-1]
        sol[i+1,:] += 0.5*(delta**2) *dot( jacob(sol[i,:],t[i]) , eval_edo )
    return sol
```

A continuaci贸n vemos un ejemplo del p茅ndulo en donde utilizamos los siguientes par谩metros:
$$
\omega _0 = 1\\
\gamma = 0.1 \\
A = 0.1 \\
\omega _1 = 0.1
$$

Se hace la gr谩fica $\theta (t)$ y $v_{\theta} (t)$ como funci贸n de tiempo.

```python
title("P茅ndulo como funci贸n del tiempo"),xlabel("t") 
tiempo = arange(0,80*pi,0.01)
solucion = Euler_mejorado_J_t(pendulo_ff,[0.5,0.0],tiempo,pendulo_ff_J)
plot(tiempo,solucion[:,0],label=r"$\theta (t)$")
plot(tiempo,solucion[:,1],label=r"$v_{\theta} (t)$")
legend(),grid()
show()
```

Como se puede apreciar, en ambas gr谩ficas, al principio se observa el amortiguamiento, la parte transitoria en donde ya no hay tanto amortiguamiento, pero empieza a dominar la frecuencia de forzamiento y finalmente se tiene que el sistema termina oscilando con la frecuencia y la amplitud del forzamiento.

A continuaci贸n se observa el espacio fase de la simulaci贸n anterior y se dibujan de manera aproximada las tres etapas que se mencionan en el p谩rrafo anterior.

```python
title("Espacio fase del p茅ndulo con fricci贸n y forzado"),xlabel("$\theta(t)$"),ylabel("$v_{\theta}(t)$")
plot(solucion[0:5000,0],solucion[0:5000,1],label="Trayectoria inicial",lw=1.5)
plot(solucion[5000:10000,0],solucion[5000:10000,1],label="Trasitorio",lw=1.5)
plot(solucion[10000:,0],solucion[10000:,1],label="Trayectoria final",lw=1.5)
gca().set_aspect("equal")
grid()
show()
```

Una modificaci贸n al algoritmo de integraci贸n es incorporar la capacidad de modificar los par谩metros que vienen por defecto en la funci贸n que queremos resolver. Tambi茅n podemos poner todos los par谩metros de las funciones en una sola variable como se muestra a continuaci贸n en las l铆neas `4`, y `12` para el caso de las funciones que representan el sistema de ecuaciones diferenciales. En el caso del algoritmo de integraci贸n, basta con hacer la evaluaci贸n del sistema de ecuaciones diferenciales `edo` y `jacob` con la variable `param`.

```python
def pendulo_ff(t_vt,t,param=[1,0.1,0.1,0.1]):
    # t_vt = [ theta , v_theta ]
    # param = [ omega 0 , gamma , A , omega 1]
    omega_0,gamma,A,omega_1=param
    dth = t_vt[1]
    dv_th = -omega_0*sin(t_vt[0]) - gamma*t_vt[1] + A*cos(omega_1*t)
    return array([dth,dv_th,1.0])

def pendulo_ff_J(t_vt,t,param=[1,0.1,0.1,0.1]):
    # t_vt = [ theta , v_theta ]
    # param = [ omega 0 , gamma , A , omega 1]
    omega_0,gamma,A,omega_1=param
    J = zeros((2,3))
    J[0,1] = 1.0
    J[1,0] = -omega_0*cos(t_vt[0])
    J[1,1] = -gamma
    J[1,2] = -A*omega_1*sin(omega_1*t)
    return J

def Euler_mejorado_J_t_p(edo,p_ini,t,jacob,param):
    # J y g con dependencia temporal y de parametros externos
    sol = zeros((len(t),len(p_ini)))
    sol[0,:] = p_ini
    delta = t[1]-t[0]
    for i in range(0,len(t)-1):
        eval_edo = edo(sol[i,:],t[i],param)
        if len(eval_edo) == len(p_ini):
            sol[i+1,:] = sol[i,:] + delta*eval_edo
        else:
            sol[i+1,:] = sol[i,:] + delta*eval_edo[0:-1]
        sol[i+1,:] += 0.5*(delta**2) *dot( jacob(sol[i,:],t[i],param) , eval_edo )
    return sol
```

De esta manera podemos utilizar la variable `param` para cambiar la frecuencia de oscilaci贸n de forzamiento. En la siguiente simulaci贸n se utilizan los mismos par谩metros que los anteriores, pero con frecuencia de forzamiento $\omega _1 =$ 0.1, 0.5 y 1. 

```python
title("P茅ndulo con diferentes frecuencias de forzamiento"),xlabel(r"$t$"),ylabel(r"$\theta(t)$")
tiempo = arange(0,60*pi,0.01)
omegas_1 = [0.1,0.5,1]
for omega_1 in  omegas_1:
    param = [1,0.05,0.1,omega_1]
    solucion = Euler_mejorado_J_t_p(pendulo_ff,[0.5,0.0],tiempo,pendulo_ff_J,param)
    plot(tiempo,solucion[:,0],lw=1.5,label="$\omega _0=$ %2.2f , $\omega_1=$ %2.2f" %(param[0],param[3] ))
legend(),grid()
show()
```

Como se puede observar, en el caso con las frecuencias de forzamiento $\omega _1 = $ 0.1 y 0.5 la amplitud final de oscilaci贸n no sobre pasa la amplitud del forzamiento, pero para $\omega _1 = 1$ se tiene que la oscilaci贸n final se ve amplificada debido a que el sistema entra en resonancia.
