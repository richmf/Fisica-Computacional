En ocasiones tenemos que el domino de la funci贸n no tiene los puntos equidistantes. Esto se debe principalmente a que en muchas ocasiones se necesita una mayor precisi贸n en representar de una mejor manera el c谩lculo de algunos puntos en la funci贸n con la que vamos a trabajar, t铆picamente se presenta en regiones donde las funciones oscilan mucho. En este *notebook* veremos como hacer una malla num茅rica en una dimensi贸n sobre la que se har谩 la evaluaci贸n de alguna funci贸n. Comencemos con importar las bibliotecas que utilizaremos, que en este caso es para graficar.

```python
from pylab import *
```

Veamos como generar algunas mallas en el intervalo $[0,2\pi]$. En la variable `x` pondremos 煤nicamente 10 puntos equidistantes, en la variable `x_2` pondremos 100 puntos equidistantes, y en la variable `x_yo` pondremos un arreglo compuesto por tres secciones, de 0 a 1 con 5 puntos, de 1.1 a 2 con 8 puntos y de 2.1 a $\pi$ 5 puntos.

```python
# malla 1
x = linspace(0,pi,10) 
# malla 2
x_2 = linspace(0,pi,100)
# malla 3
x_a = linspace(0,1,5)
x_b = linspace(1.1,2,8) 
x_c = linspace(2.1,pi,5)
x_yo = concatenate((x_a,x_b,x_c)) #malla compuesta
# evaluacion de la funcion para cada malla
y = sin(3*x)
y_2 = sin(3*x_2)
y_yo = sin(3*x_yo)
fig, (ax1,ax2,ax3) = subplots(nrows=3,ncols=1,dpi=100)
ax1.plot(x,y,ls='dashed',marker='o')
ax1.set_ylim([-1.1,1.1]) , ax1.set_title("sin(3x) con 10 pts.")
ax2.plot(x_2,y_2,ls='dashed',marker='*')
ax2.set_ylim([-1.1,1.1]) , ax2.set_title("sin(3x) con 100 pts.")
ax3.plot(x_yo,y_yo,ls='dashed',marker='*')
ax3.set_ylim([-1.1,1.1]) , ax3.set_title("sin(3x) con malla compuesta")
tight_layout()
```

Como se puede apreciar, entre m谩s puntos se tengan en el arreglo num茅rico o malla en 1D, la forma de la gr谩fica tiene m谩s informaci贸n de la funci贸n. En el caso con 100 puntos se pueden observar todas las oscilaciones de la funci贸n, mientras que con la malla de 10 puntos no se aprecian bien las oscilaciones. Sin embargo, se puede obtener un resultado razonable con la 煤ltima malla compuesta que recupere las oscilaciones de una mejor manera, ya que se puede apreciar que esta malla si alcanza los m谩ximos y m铆nimos que tiene la funci贸n. Esta malla compuesta tiene 18 puntos que son mucho menos que 100 y no mucho m谩s que 10.

Otra manera de hacer una malla, es hacerla de forma funcional en el dominio. Para ello podemos hacer el mapeo de alg煤n intervalo y ajustarlo al que deseemos utilizar. En el caso de abajo elevamos los puntos el intervalo $[0,2\pi]$ a la quinta potencial y posteriormente dividimos por $\pi ^4$ para que al final nos quede en el intervalo $[0,2\pi]$. Esto acomoda los puntos en el intervalo de tal manera que 茅stos se van espaciando como $x^5$.

```python
x2 = (x**5)*(1/pi**4)
y_x2 = sin(3*x2)
figure(dpi=100)
plot(x2,y_x2,ls='dashed',marker='o',label="Malla espaciada con x/")
plot(x_2,y_2,label="sin(x)")
legend(loc=1)
show()
```

Otra opci贸n es generar una malla con puntos aleatorios dentro del intervalo deseado. En el caso de abajo se generan 30 puntos aleatorios en el intervalo $[0,1]$, se multiplican por $\pi$ para que queden en el intervalo que queremos y finalmente se ordenan de menor a mayor para que se puedan evaluar por la funci贸n que uno quiera.

```python
x_random = pi*sort(rand(30))
y_random = sin(3*x_random)
figure(dpi=100)
plot(x_2,y_2,label="sin(x)")
plot(x_random,y_random,ls='dashed',marker='o',label="Malla aleatoria")
legend()
show()
```

Como se puede apreciar, los puntos quedan distribuidos aleatoriamente dentro del intervalo $[0,\pi]$.
