En c贸mputo cient铆fico es muy importante saber muchos par谩metros de nuestros experimentos num茅ricos. El objetivo de este notebook es saber como podemos calcular dichos tiempos de ejecuci贸n. La biblioteca que utilizaremos para medir los tiempos es `time`:

```python
from pylab import *
import time
```

Supongamos que queremos medir el tiempo que tarda en ejecutarse la funci贸n exponencial, $e^x$, sobre un arreglo num茅rico. Para ello utilizaremos la biblioteca `time`, misma que nos proporciona marcas de tiempo con el que podemos hacer los c谩lculos. En la siguiente celda se muestra el c谩lculo de dicho ejemplo

```python
x = linspace(0,10,10)
inicio = time.time()
exp(x)
fin = time.time()
t_i = fin - inicio
print("Tiempo de ejecuci贸n: ",t_i)
```

En el caso anterior s贸lo hemos calculado el tiempo que tarde en ejecutarse la exponencial sobre dicho arreglo. Para tener una mejor medici贸n podemos utilizar un ciclo `for` y hacer varias veces el c谩lculo para que al final podamos obtener una medici贸n promedio.

```python
x = linspace(0,10,10)
n_experimentos = 100000
t_is =  zeros(n_experimentos)
for i in range(n_experimentos):
    inicio = time.time()
    exp(x)
    fin = time.time()
    t_is[i] = fin - inicio
print("Tiempo de ejecuci贸n: ",mean(t_is))
```

Veamos ahora que pasa si hacemos el experimento num茅rico para diferente n煤mero de puntos en el dominio de la funci贸n, es decir, en este caso veamos que pasa cuando el n煤mero de puntos se incrementa de 10 en 10 hasta llegar a 100. Para lo anterior hagamos un ciclo que nos ayude a calcular dichos tiempos y guardarlos en la variable `tiempos`.

```python
num_puntos = linspace(10,100,10)
tiempos = zeros(len(num_puntos))
for n_i in range(len(num_puntos)):
    x = linspace(0,10,int(num_puntos[n_i]))
    n_experimentos = 100000
    t_is = zeros(n_experimentos)
    for i in range(n_experimentos):
        inicio = time.time()
        exp(x)
        fin = time.time()
        t_is[i] = fin-inicio
    tiempos[n_i] = mean(t_is)
```

```python
figure(dpi=100)
title("Tiempos promedio para evaluar $e^x$")
xlabel("N煤mero de puntos")
ylabel("Tiempo en [$\mu s$]")
plot(num_puntos,tiempos*1e6,'o-')
grid()
show()
```

Como se puede apreciar, en este caso particular,  el tiempo se incrementa linealmente conforme aumentamos los puntos en nuestra malla 1D en el c谩lculo de la funci贸n.
