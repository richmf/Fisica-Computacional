En cómputo científico es muy importante saber muchos parámetros de nuestros experimentos numéricos. El objetivo de este notebook es saber como podemos calcular dichos tiempos de ejecución. La biblioteca que utilizaremos para medir los tiempos es `time`:

```python
from pylab import *
import time
```

Supongamos que queremos medir el tiempo que tarda en ejecutarse la función exponencial, $e^x$, sobre un arreglo numérico. Para ello utilizaremos la biblioteca `time`, misma que nos proporciona marcas de tiempo con el que podemos hacer los cálculos. En la siguiente celda se muestra el cálculo de dicho ejemplo

```python
x = linspace(0,10,10)
inicio = time.time()
exp(x)
fin = time.time()
t_i = fin - inicio
print("Tiempo de ejecución: ",t_i)
```

En el caso anterior sólo hemos calculado el tiempo que tarde en ejecutarse la exponencial sobre dicho arreglo. Para tener una mejor medición podemos utilizar un ciclo `for` y hacer varias veces el cálculo para que al final podamos obtener una medición promedio.

```python
x = linspace(0,10,10)
n_experimentos = 100000
t_is =  zeros(n_experimentos)
for i in range(n_experimentos):
    inicio = time.time()
    exp(x)
    fin = time.time()
    t_is[i] = fin - inicio
print("Tiempo de ejecución: ",mean(t_is))
```

Veamos ahora que pasa si hacemos el experimento numérico para diferente número de puntos en el dominio de la función, es decir, en este caso veamos que pasa cuando el número de puntos se incrementa de 10 en 10 hasta llegar a 100. Para lo anterior hagamos un ciclo que nos ayude a calcular dichos tiempos y guardarlos en la variable `tiempos`.

```python
num_puntos = linspace(10,100,10)
tiempos = zeros(len(num_puntos))
for n_i in range(len(num_puntos)):
    x = linspace(0,10,int(num_puntos[n_i]))
    n_experimentos = 100000
    t_is = zeros(n_experimentos)
    for i in range(n_experimentos):
        inicio = time.time()
        exp(x)
        fin = time.time()
        t_is[i] = fin-inicio
    tiempos[n_i] = mean(t_is)
```

```python
figure(dpi=100)
title("Tiempos promedio para evaluar $e^x$")
xlabel("Número de puntos")
ylabel("Tiempo en [$\mu s$]")
plot(num_puntos,tiempos*1e6,'o-')
grid()
show()
```

Como se puede apreciar, en este caso particular,  el tiempo se incrementa linealmente conforme aumentamos los puntos en nuestra malla 1D en el cálculo de la función.

> Nota: Como se puede observar los tiempos son ligeramente mayores en **Python** en comparación con los tiempos que se obtienen en **Julia** (versión equivalente de este notebook).
